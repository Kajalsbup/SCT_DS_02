```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Course Name': ['Python Programming', 'Data Science', 'Machine Learning', 'Web Development', 'Cyber Security',
                     'Java Programming', 'Cloud Computing', 'Artificial Intelligence', 'Network Security', 'Database Management'],
    'Price': [199.99, 299.99, 399.99, 149.99, 249.99, 179.99, 349.99, 449.99, 199.99, 299.99],
    'Duration': [30, 40, 50, 20, 30, 25, 45, 55, 35, 40],
    'Category': ['Programming', 'Data Science', 'AI', 'Web Development', 'Cyber Security', 'Programming', 'Cloud Computing', 'AI', 'Cyber Security', 'Database Management'],
    'Rating': [4.5, 4.8, 4.9, 4.2, 4.6, 4.3, 4.7, 4.9, 4.5, 4.6],
    'Enrollments': [1000, 800, 1200, 900, 1100, 900, 1000, 1300, 1000, 1200],
    'Level': ['Beginner', 'Intermediate', 'Advanced', 'Beginner', 'Intermediate', 'Beginner', 'Intermediate', 'Advanced', 'Intermediate', 'Advanced'],
    'Instructor': ['John', 'Jane', 'Bob', 'Alice', 'Mike', 'John', 'Jane', 'Bob', 'Alice', 'Mike']
}
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
#Data Cleaning
# Check for missing values
print("Missing Values:")
print(df.isnull().sum())
```

```python
# Drop unnecessary columns
df = df.drop(['Course Name'], axis=1)
```

```python
# Scale the data using StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[['Price', 'Duration', 'Rating', 'Enrollments']] = scaler.fit_transform(df[['Price', 'Duration', 'Rating', 'Enrollments']])

# Print the updated DataFrame
print("Updated DataFrame:")
print(df)
```

```python
print("Summary Statistics:")
print(df.describe())
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Category', y='Price', data=df)
plt.title('Boxplot of Price by Category')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a scatter plot of the Price column vs the Duration column
plt.figure(figsize=(10,6))
sns.scatterplot(x='Duration', y='Price', data=df)
plt.title('Scatterplot of Price vs Duration')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a histogram of the Price column
plt.figure(figsize=(10,6))
sns.histplot(df['Price'], bins=10, kde=True)
plt.title('Histogram of Price')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a bar plot of the mean Price by Category
plt.figure(figsize=(10,6))
sns.barplot(x='Category', y='Price', data=df, estimator=np.mean)
plt.title('Barplot of Mean Price by Category')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a line plot of the Price column over the Duration column
plt.figure(figsize=(10,6))
sns.lineplot(x='Duration', y='Price', data=df)
plt.title('Lineplot of Price over Duration')
plt.xlabel('Duration')
plt.ylabel('Price')
plt.legend()
plt.show()
```
